```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
import csv
import pandas as pd
```

Load in the data we want (with only the first 13 columns of the predictions metadata because I wrote them to the file twice (oops))

```python
model_history = pd.read_csv("../ml_model_data/classification_model/history.csv", index_col=0, header=None).T.to_xarray()
```

Pull out the variable fields we want

```python
loss = model_history.loss.values
accuracy = model_history.accuracy.values
```

Plot the loss function during training

```python
plt.plot(loss)
plt.xlabel("Epochs")
plt.ylabel("Categorical Cross Entropy")
plt.yscale('log')
plt.xscale('log')
plt.title("Classification Model Loss During Training")
plt.savefig("training_loss_100_5000")
plt.draw()
```

Plot the accuracy during training

```python
plt.plot(accuracy)
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Classification Model Accuracy During Training")
plt.savefig("training_accuracy_20210716")
plt.draw()
```
