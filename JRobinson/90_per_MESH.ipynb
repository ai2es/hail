```python
f_cs = "c:/Users/jorda/hagelslag/covnet_output/MESH_20160413_03z.nc"
data = netCDF4.Dataset(f_cs, "r")
```

```python
longitude = data.variables['longitude'][:].data
latitude = data.variables['latitude'][:].data
MESH = data.variables['unknown'][:].data
quantile = data.variables['quantile'][:].data
```

```python
crs = ccrs.LambertConformal(central_longitude=-100.0, central_latitude=45.0)

# Create the figure and plot background on different axes

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(20, 13), constrained_layout=True,
                          subplot_kw={'projection': crs})

ax.set_extent([230., 290., 20., 48.])
ax.add_feature(cfeature.COASTLINE.with_scale('50m'), linewidth=0.5)
ax.add_feature(cfeature.STATES, linewidth=0.5)
ax.add_feature(cfeature.BORDERS, linewidth=0.5)

ax.pcolormesh(longitude,latitude,MESH,vmin=0,cmap='turbo',transform=ccrs.PlateCarree())
plt.title('MESH 0250 UTC', fontsize='medium')

plt.savefig('c:/Users/jorda/Downloads/90_MESH_0.png',dpi=300)
```
