```python
import glob
files = glob.glob("HRRR_data_2019_05_16/*.grib2")
files.sort()
file_list = []
file_list_2 = []
for file in files:
    file_list.append(xr.open_dataset(file, filter_by_keys={'typeOfLevel': 'atmosphere'}))
    file_list_2.append(xr.open_dataset(file, filter_by_keys={'stepType': 'instant', 'typeOfLevel': 'surface'}))
ds_i = xr.concat(file_list, dim="time")
ds_o = xr.concat(file_list_2, dim="time")
```

```python
ref = ds_i.refc.isel(time=slice(0,23))
cape = ds_o.cape.isel(time=slice(0,23))
ref.shape
```

```python
ref.to_netcdf("unet_test_data/HRRR_refc_0000_2200.nc")
cape.to_netcdf("unet_test_data/HRRR_cape_0000_2200.nc")
```
